```python
import pandas as pd
from matplotlib import pyplot as plt
import csv
```

```python
startships_counter= {}

with open('starwars/csv/starships.csv','r') as file:
    reader = csv.DictReader(file)

    for row in reader:
        length = row['length']
        if length in startships_counter:
            startships_counter[length]+=1

        else:
            startships_counter[length]=1


startships_label = list(startships_counter.keys())
starships_values = list(startships_counter.values())

labels_num = len(startships_label)
colors = plt.cm.get_cmap('tab10', labels_num)
plt.bar(startships_label,starships_values,color=colors(range(labels_num)))
plt.xlabel('Longitud')
plt.ylabel('Naves')
plt.title('Longitud de naves')
plt.xticks(rotation=90)
plt.xlim(-0.5, len(startships_label)-0.5)
plt.tick_params(axis='x',which='major',labelsize=8)
plt.show()
```
