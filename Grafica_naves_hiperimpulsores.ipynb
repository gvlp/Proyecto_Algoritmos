```python
import pandas as pd
from matplotlib import pyplot as plt
import csv
```

```python
data = pd.read_csv('starwars/csv/starships.csv')

starships = data['name']
lenghs = data['hyperdrive_rating']

labels_num = len(starships)
colors = plt.cm.get_cmap('tab10', labels_num)
plt.bar(starships,lenghs,color=colors(range(labels_num)))
plt.xlabel('Naves')
plt.ylabel('Capacidad de carga')
plt.title('Comparacion de capacidad de carga de naves')
plt.xticks(rotation=90)
plt.xlim(-0.5, len(starships)-0.5)
plt.tick_params(axis='x',which='major',labelsize=7)
plt.show()
```
